```python
from bs4 import BeautifulSoup
import requests

base_url = 'https://leroymerlin.ru/catalogue/sadovaya-mebel/'
all_urls = []
all_img_urls = []

response = requests.get(base_url, cookies=cookies, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

pagination = soup.find('div', {'class': 's1nb31eh_plp'})
num_pages = int(pagination.find_all('a', {'class': 'bex6mjh_plp'})[-2].text)

for i in range(1, num_pages + 1):
    all_urls.append(base_url + '?page=' + str(i))

print(f"Number of pages: {num_pages}")
print("URLs of all pages:")
for url in all_urls:
    print(url)
```

```python
from time import sleep

for url in all_urls:
    response = requests.get(url, cookies=cookies, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    img_tags = soup.find_all('img')
    img_urls = [img['src'] for img in img_tags if 'src' in img.attrs]
    all_img_urls.extend(img_urls)
    print(img_urls)
    sleep(3)
```

```python
import os
import requests

os.makedirs('images/test', exist_ok=True)

for url in all_img_urls:
    filename = url.split('/')[-1]
    try:
        with open(f'images/test/{filename}', 'wb') as f:
            f.write(requests.get(url).content)
    except:
        print(f'Error downloading {url}')
    sleep(0.1)

print('Images downloaded successfully')
```
